```python
##metropolis yakhzade
import numpy as np
import matplotlib.pyplot as plt
f=lambda x,mean,sigma:(1/(sigma*np.sqrt(2*np.pi)))*np.exp(-(1/2)*((x-mean)**2)*1/sigma**2)
mean=0
sigma=3
x=mean
n=10**5
a=[]
i=0
while i<n:
    y=np.random.uniform(-10,10)
    px=f(x,mean,sigma)
    py=f(y,mean,sigma)
    w=min(1,py/px)
    r=np.random.uniform(0,1)
    if r<w:
        x=y
        a.append(x)
        i+=1
plt.hist(a,bins=40)
plt.show()
```

```python
#metropolis
##metropolis yakhzade
import numpy as np
import matplotlib.pyplot as plt
f=lambda x,mean,sigma:(1/(sigma*np.sqrt(2*np.pi)))*np.exp(-(1/2)*((x-mean)**2)*1/sigma**2)
mean=0
sigma=0.4
x=mean
n=2*10**5
a=[]
#ar={0.2:0.9, 0.4:0.8, 0.63:0.7, 0.86:0.6, 1.15:0.5, 1.55:0.4, 2.1:0.3, 3.2:0.2, 6:0.1}
alpha=6
accept=0
for i in range(n):
    y=x+alpha*np.random.uniform(-1,1)
    px=f(x,mean,sigma)
    py=f(y,mean,sigma)
    w=min(1,py/px)
    r=np.random.uniform(0,1)
    if r<w:
        x=y
        a.append(x)
        accept+=1
print(accept/n)
#plt.hist(a,bins=40)
#plt.show()
j=np.arange(0,len(a)//10**3,1)
cor=[]
for i in j:
    x=a[0:len(a)-i]
    y=a[i:len(a)]   
    cor.append(np.corrcoef(x,y)[0,1])
for count,item in enumerate(cor):
    if item<=1/np.exp(1):
        print(count)
        break

plt.plot(j,cor)
plt.show()
```
